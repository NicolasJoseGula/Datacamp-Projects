# Planning a Trip to Paris with the OpenAI API

A conversational AI travel guide for Paris, built using the OpenAI Chat Completions API. The assistant acts as a virtual Parisian expert, answering tourist questions while maintaining full conversation history across multiple turns.

## Setup

Load environment variables and initialize the OpenAI client. The API key is read from a `.env` file to keep credentials out of the code.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')

model = "gpt-4o-mini"
client = OpenAI(api_key=OPENAI_API_KEY)
```

## System Prompt

The system prompt defines the assistant's persona and behavior before the conversation starts. It instructs the model to act as a knowledgeable Parisian travel expert — scoping its responses to the domain and setting the tone for the entire conversation.

```python
messages = [
    {
        "role": "system",
        "content": (
            "You are a virtual Parisian expert, delivering valuable insights into the city's "
            "iconic landmarks and hidden treasures. You will respond intelligently to a set of "
            "common questions, providing a more engaging and immersive travel planning experience "
            "for the clientele of Peterman Reality Tours. Your answers are precise but concise."
        )
    }
]
```

## Conversation Loop

Each question is sent to the API along with the full conversation history. This is the key mechanism that gives the assistant memory — every API call includes all previous messages, so the model has full context when generating each response.

After each response, the assistant's reply is appended to `messages`, growing the history for the next turn.

```python
conversation = [
    "How far away is the Louvre from the Eiffel Tower (in miles) if you are driving?",
    "Where is the Arc de Triomphe?",
    "What are the must-see artworks at the Louvre Museum?",
]

for q in conversation:
    print("[USER]: ", q)

    user_dict = {"role": "user", "content": q}
    messages.append(user_dict)

    response = client.chat.completions.create(
        model=model,
        messages=messages,
        max_completion_tokens=1000,
        temperature=0.5
    )

    assistant_response = {"role": "assistant", "content": response.choices[0].message.content}
    messages.append(assistant_response)
    print("[ASSISTANT]: ", response.choices[0].message.content, "\n")
```
